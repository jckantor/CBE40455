# Economic Order Quantity

This [Jupyter notebook](http://jupyter.org/notebook.html) introduces and illustrates the concept of Economic Order Quantity.

## Initializations

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context('talk')
```

## Economic Order Quantity (EOQ)

```python
K = 200.00      # Cost to place an order
D = 1200.       # Annual Demand
h = 25.00       # Cost of holding an item for one year

HC = lambda Q: Q*h/2.0
OC = lambda Q: K*D/Q
AC = lambda Q: OC(Q) + HC(Q)

EOQ = np.sqrt(2.0*D*K/h)

plt.figure(figsize=(12,8))
Q = np.linspace(50,400)
plt.plot(Q,AC(Q),lw=2)
plt.xlim(0,400)
plt.ylim(0,plt.ylim()[1])

plt.plot(Q,HC(Q),'r',lw = 2)
plt.plot(Q,OC(Q),'g',lw = 2)
plt.plot([EOQ,EOQ,0],[0,AC(EOQ),AC(EOQ)],'k--')

plt.xlabel('Q: Order Quantity')
plt.ylabel('Annual Cost')
plt.title('Variable Costs of Maintaining Inventory')

aprops = dict(shrink=0.1,width=0.5,headwidth=6,headlength=8)

plt.text(240,4600,'Total Inventory Costs',horizontalalignment='right')
plt.annotate('', xy=(275,AC(275)), xytext=(240,4600), arrowprops=aprops)

plt.text(250,2500,'Holding Costs = $h Q/2$',verticalalignment='center')
plt.annotate('', xy=(225,HC(225)), xytext=(250,2500), arrowprops=aprops)

plt.text(250,1500,'Order Costs = $K D/Q$',verticalalignment='center')
plt.annotate('', xy=(225,OC(225)), xytext=(250,1500), arrowprops=aprops)

plt.text(200,300,'Economic Order Quantity',verticalalignment='center')
plt.annotate('', xy=(EOQ,0), xytext=(200,300), arrowprops=aprops)

plt.text(25,1900,'Minimum Cost',verticalalignment='center')
plt.annotate('', xy=(0,AC(EOQ)), xytext=(25,1900), arrowprops=aprops)

plt.savefig('../img/EconomicOrderQuantity.png')
```
